```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns",None)
```

```python
master = pd.read_csv("master_healthcare_dataset.csv")

nfhs = pd.read_excel("NFHS_5_India_Districts_Factsheet_Data.xls")
```

```python
def clean_columns(df):

    df.columns = (
        df.columns
        .str.lower()
        .str.strip()
        .str.replace(" ","_")
        .str.replace("-","_")
        .str.replace("/","_")
    )

    return df

nfhs = clean_columns(nfhs)
```

```python
master["state"] = master["state"].str.upper().str.strip()
master["district"] = master["district"].str.upper().str.strip()

nfhs["state_ut"] = nfhs["state_ut"].str.upper().str.strip()
nfhs["district_names"] = nfhs["district_names"].str.upper().str.strip()
```

```python
nfhs["state_ut"] = nfhs["state_ut"].replace({

    "NCT OF DELHI":"DELHI",

    "MAHARASTRA":"MAHARASHTRA",

    "ODISHA":"ORISSA",

    "ANDAMAN & NICOBAR ISLANDS":"ANDAMAN AND NICOBAR ISLANDS",

    "JAMMU & KASHMIR":"JAMMU AND KASHMIR"

})
```

```python
nfhs = nfhs[[

"state_ut",

"district_names",

"population_living_in_households_with_electricity_(%)",

"households_using_clean_fuel_for_cooking3_(%)",

"households_with_any_usual_member_covered_under_a_health_insurance_financing_scheme_(%)",

"women_(age_15_49)_who_are_literate4_(%)",

"mothers_who_had_at_least_4_antenatal_care_visits__(for_last_birth_in_the_5_years_before_the_survey)_(%)",

"institutional_births_(in_the_5_years_before_the_survey)_(%)",

"children_age_12_23_months_fully_vaccinated_based_on_information_from_either_vaccination_card_or_mother's_recall11_(%)",

"children_under_5_years_who_are_stunted_(height_for_age)18_(%)",

"children_under_5_years_who_are_underweight_(weight_for_age)18_(%)",

"all_women_age_15_49_years_who_are_anaemic22_(%)"

]]
```

```python
nfhs.rename(columns={

"state_ut":"state",

"district_names":"district",

"population_living_in_households_with_electricity_(%)":"electricity",

"households_using_clean_fuel_for_cooking3_(%)":"clean_fuel",

"households_with_any_usual_member_covered_under_a_health_insurance_financing_scheme_(%)":"health_insurance",

"women_(age_15_49)_who_are_literate4_(%)":"female_literacy",

"mothers_who_had_at_least_4_antenatal_care_visits__(for_last_birth_in_the_5_years_before_the_survey)_(%)":"antenatal_care",

"institutional_births_(in_the_5_years_before_the_survey)_(%)":"institutional_birth",

"children_age_12_23_months_fully_vaccinated_based_on_information_from_either_vaccination_card_or_mother's_recall11_(%)":"vaccination",

"children_under_5_years_who_are_stunted_(height_for_age)18_(%)":"stunted",

"children_under_5_years_who_are_underweight_(weight_for_age)18_(%)":"underweight",

"all_women_age_15_49_years_who_are_anaemic22_(%)":"anaemia"

},inplace=True)
```

```python
final_master = master.merge(

nfhs,

on=["state","district"],

how="left"

)
```

```python
print(final_master.shape)

print(final_master.isnull().sum())
```

```python
numeric = final_master.select_dtypes(include=np.number).columns

final_master[numeric] = final_master[numeric].fillna(

final_master[numeric].median()

)
```

```python
final_master.to_csv(

"final_healthcare_master.csv",

index=False

)

print("Dataset Saved")
```
