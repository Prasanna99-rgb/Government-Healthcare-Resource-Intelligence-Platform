```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns",None)
```

```python
df = pd.read_csv("healthcare_priority_index.csv")

df.head()
```

```python
def generate_recommendation(row):

    recommendations = []

    if row["population_per_health_centre"] > 30000:
        recommendations.append("Build New PHCs")

    if row["health_centres"] < 5:
        recommendations.append("Recruit Additional Doctors & Nurses")

    if row["vaccination"] < 70:
        recommendations.append("Launch Vaccination Campaign")

    if row["anaemia"] > 50:
        recommendations.append("Start Nutrition & Anaemia Control Programme")

    if row["female_literacy"] < 70:
        recommendations.append("Women's Health Awareness Programme")

    if row["antenatal_care"] < 60:
        recommendations.append("Improve Maternal Healthcare Services")

    if row["institutional_birth"] < 70:
        recommendations.append("Strengthen Institutional Delivery Facilities")

    if row["stunted"] > 35:
        recommendations.append("Child Nutrition Improvement Programme")

    if row["underweight"] > 30:
        recommendations.append("School Nutrition Initiative")

    if len(recommendations)==0:
        recommendations.append("Current Healthcare Infrastructure is Satisfactory")

    return " | ".join(recommendations)
```

```python
df["government_recommendation"] = df.apply(
    generate_recommendation,
    axis=1
)
```

```python
df[[
    "district",
    "state",
    "priority_level",
    "government_recommendation"
]].head(20)
```

```python
recommendation_counts = (
    df["government_recommendation"]
    .str.split(" \\| ")
    .explode()
    .value_counts()
    .reset_index()
)

recommendation_counts.columns = [
    "Recommendation",
    "Count"
]

recommendation_counts
```

```python
import plotly.express as px

fig = px.bar(
    recommendation_counts,
    x="Recommendation",
    y="Count",
    color="Count",
    title="Government Policy Recommendation Distribution"
)

fig.show()
```
