```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns",None)
```

```python
df = pd.read_csv("final_healthcare_master.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print("Total States :",df["state"].nunique())

print("Total Districts :",df["district"].nunique())

print("Total Population :",df["population"].sum())

print("Total Health Centres :",df["health_centres"].sum())

print("Average Literacy :",round(df["literacy"].mean(),2))

df["vaccination"] = pd.to_numeric(df["vaccination"], errors='coerce')
print("Average Vaccination :",round(df["vaccination"].mean(),2))

print("Average Anaemia :",round(df["anaemia"].mean(),2))
```

```python
top=df.sort_values("population",ascending=False).head(20)

fig=px.bar(
top,
x="district",
y="population",
color="population",
title="Top 20 Most Populated Districts"
)

fig.show()
```

```python
state_pop=df.groupby("state")["population"].sum().reset_index()

fig=px.bar(

state_pop.sort_values("population",ascending=False),

x="state",

y="population",

title="State-wise Population"

)

fig.show()
```

```python
state_health=df.groupby("state")["health_centres"].sum().reset_index()

fig=px.bar(

state_health.sort_values("health_centres",ascending=False),

x="state",

y="health_centres",

color="health_centres",

title="Health Centres by State"

)

fig.show()
```

```python
top=df.sort_values(

"population_per_health_centre",

ascending=False

).head(20)

fig=px.bar(

top,

x="district",

y="population_per_health_centre",

color="population_per_health_centre",

title="Highest Healthcare Pressure"

)

fig.show()
```

```python
lowest=df.sort_values("vaccination").head(20)

fig=px.bar(

lowest,

x="district",

y="vaccination",

color="vaccination",

title="Lowest Vaccination Districts"

)

fig.show()
```

```python
vacc=df.groupby("state")["vaccination"].mean().reset_index()

fig=px.bar(

vacc.sort_values("vaccination",ascending=False),

x="state",

y="vaccination",

title="Average Vaccination by State"

)

fig.show()
```

```python
anaemia=df.sort_values("anaemia",ascending=False).head(20)

fig=px.bar(

anaemia,

x="district",

y="anaemia",

color="anaemia",

title="Highest Anaemia Districts"

)

fig.show()
```

```python
birth=df.groupby("state")["institutional_birth"].mean().reset_index()

fig=px.bar(

birth.sort_values("institutional_birth"),

x="state",

y="institutional_birth",

title="Institutional Birth Rate"

)

fig.show()
```

```python
antenatal=df.groupby("state")["antenatal_care"].mean().reset_index()

fig=px.bar(

antenatal.sort_values("antenatal_care"),

x="state",

y="antenatal_care",

title="Antenatal Care Coverage"

)

fig.show()
```

```python
stunted=df.sort_values("stunted",ascending=False).head(20)

fig=px.bar(

stunted,

x="district",

y="stunted",

title="Highest Child Stunting"

)

fig.show()
```

```python
under=df.sort_values("underweight",ascending=False).head(20)

fig=px.bar(

under,

x="district",

y="underweight",

title="Highest Underweight Children"

)

fig.show()
```

```python
female=df.groupby("state")["female_literacy"].mean().reset_index()

fig=px.bar(

female.sort_values("female_literacy"),

x="state",

y="female_literacy",

title="Female Literacy"

)

fig.show()
```

```python
corr=df.select_dtypes(include=np.number).corr()

fig=px.imshow(

corr,

text_auto=True,

title="Correlation Matrix"

)

fig.show()
```
